```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# arquivo a

arq1 = np.load('ex1_d.npz')
x = arq1['x']
y = arq1['y']

id = np.argsort(x)
x = x[id]
y = y[id]

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()



"""
como o plot mostrou pontos que se assemelham a uma reta, o grau será 1
"""

A = np.ones((50, 2))

A[:, 1] = x

coef = np.linalg.lstsq(A, y, rcond= 0)[0]

f = lambda x: coef[0] + x * coef[1]

f = np.vectorize(f)

plt.plot(x, y, 'ro')
plt.plot(x, f(x), label='reta')
plt.legend()
plt.show()
```

```python
# arquivo b

arq2 = np.load('ex1_b.npz')
x = arq2['x']
y = arq2['y']

# reordena mantendo os indices (até pq cada x tem um y equivalente)
id = np.argsort(x)
x = x[id]
y = y[id]

plt.scatter(x, y)
plt.show()

"""
como o plot mostrou pontos que se assemelham a uma parábelo, o grau será 2
"""

A = np.ones((50, 3))

""" 

A:
coluna 0: só tem números 1
coluna 1: x
coluna 2: x ** 2

"""


A[:, 1] = x
A[:, 2] = x ** 2

coef = np.linalg.lstsq(A, y, rcond= 0)[0]

f = lambda x: coef[0] + x * coef[1] + x**2 * coef[2]

f = np.vectorize(f)

plt.plot(x, y, 'ro')
plt.plot(x, f(x), label='parabola')
plt.legend()
plt.show()
```

```python
# arquivo c

arq3 = np.load('ex1_c.npz')
x = arq3['x']
y = arq3['y']

id = np.argsort(x)
x = x[id]
y = y[id]


plt.scatter(x, y)
plt.show()

"""
log 
"""

A = np.ones((50, 2))

A[:, 1] = np.log10(x)

coef = np.linalg.lstsq(A, y, rcond= 0)[0]

f = lambda x: coef[0] + np.log10(x) * coef[1]

plt.plot(x, y, 'ro')
plt.plot(x, f(x), label='log')
plt.legend()
```
